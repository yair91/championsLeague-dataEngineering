```python
goles_por_equipo = df_combined_filled.groupby('Equipo')['Goles'].sum()

top_10_equipos = goles_por_equipo.nlargest(10)

plt.figure(figsize=(10, 6))
top_10_equipos.plot(kind='bar', color='skyblue')
plt.xlabel('Equipo')
plt.ylabel('Goles')
plt.title('Top 10 Equipos con Más Goles')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

for equipo, goles in top_10_equipos.items():
    print(f"{equipo}: {goles} goles")
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame
df_ataque = pd.read_csv('Kaggle/Champions_stats/attacking.csv')
df_attempts = pd.read_csv('Kaggle/Champions_stats/attempts.csv')
df_defensa = pd.read_csv('Kaggle/Champions_stats/defending.csv')
df_disciplina= pd.read_csv('Kaggle/Champions_stats/disciplinary.csv')
df_distribución = pd.read_csv('Kaggle/Champions_stats/distributon.csv')
df_portero = pd.read_csv('Kaggle/Champions_stats/goalkeeping.csv')
df_goles = pd.read_csv('Kaggle/Champions_stats/goals.csv')
df_keystats= pd.read_csv('Kaggle/Champions_stats/key_stats.csv')

# Cargar el archivo CSV en un DataFrame
df_combined_filled = pd.concat([df_ataque, df_attempts, df_defensa, df_disciplina, df_distribución, df_portero, df_goles, df_keystats]).fillna(0)

# Crear la interfaz gráfica con Streamlit
def main():
    st.title('Análisis Estadístico de la Champions League')
    st.sidebar.title('Menú')

    # Opciones del menú
    options = ['Top 10 Clubs Goleadores', 'Top 10 Goleadores']
    choice = st.sidebar.selectbox('Seleccione una opción', options)

    if choice == 'Top 10 Clubs Goleadores':
        top_10_clubs_goleadores(df_combined_filled)
    elif choice == 'Top 10 Goleadores':
        top_10_goleadores(df_combined_filled)

# Función para mostrar el top 10 de clubs goleadores
def top_10_clubs_goleadores(df):
    goles_por_equipo = df.groupby('Equipo')['Goles'].sum().nlargest(10)
    
    st.subheader('Top 10 Clubs Goleadores')
    st.bar_chart(goles_por_equipo)

    st.write('Cantidad de goles por club:')
    for equipo, goles in goles_por_equipo.items():
        st.write(f"{equipo}: {goles} goles")

# Función para mostrar el top 10 de goleadores
def top_10_goleadores(df):
    goles_por_jugador = df.groupby('Nombre Jugador')['Goles'].sum().nlargest(10)
    
    st.subheader('Top 10 Goleadores')
    st.bar_chart(goles_por_jugador)

    st.write('Cantidad de goles por jugador:')
    for jugador, goles in goles_por_jugador.items():
        st.write(f"{jugador}: {goles} goles")

if __name__ == '__main__':
    main()
```
